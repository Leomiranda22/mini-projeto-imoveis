```python
# Quantos imóveis existem na base de dados?
dados.value_counts('Tipo')\
    .sum()

# Resposta: Existem 32960 imóveis na nossa base de dados
```

```python
# Qual é o bairro com a maior quantidade de imóveis cadastrados?
imoveis_por_bairro = dados['Bairro'].value_counts()
imoveis_por_bairro.head(1)

# Resposta: Bairro da Tijuca vence com a maior quantidade de imóveis cadastrados na base com 6259 imóveis
```

```python
valor_medio = (
    dados.groupby('Tipo')['Valor']\
     .mean()\
     .reset_index()\
     .round(2)
)
valor_medio
```

```python
# Qual é o imóvel mais caro e qual é o mais barato?
dados.sort_values(by='Valor', ascending=False).head(1)

# Resposta: O imóvel mais caro é Prédio Inteiro no bairro Jacarepaguá com o valor de 120000000.0
```

```python
dados.sort_values(by='Valor', ascending=True).head(1)

# Resposta: O imóvel mais barato é Galpão/Depósito/Armazém no bairro Taquara com o valor de 75.0
```
